```python
x = "awesome"

def myfunc():
  print("Python is " + x)
  print("partho")

myfunc()
```

local variable

```python
x = "awesome"

def myfunc():
  # local variable x is prioratized while executing function
  x = "fantastic"
  print("Python is " + x)

myfunc()

print("Python is " + x) # x is global variable
```

'global' keyword = to change a global variable inside a function.


If we assign a value to a variable inside the function, python considers it to be a local variable. That's why we have to specify 'global' to refer to the global variable.

```python
x = "awesome"
print("Python is " + x)

def myfunc():
  # changing global varible inside a function
  global x
  x = "fantastic"
  print("Python is " + x)

myfunc()

print("Python is " + x) #  x is global variable
```

# Data types

type() = getting the data type

```python
x = 5
print(type(x))
```

```python
#str
x = "Hello World"
print(x)
print(type(x))
print("\n")

#int
x = 20
print(x)
print(type(x))
print("\n")

#float
x = 20.5
print(x)
print(type(x))
print("\n")

#complex
x = 2j
print(x)
print(type(x))
print("\n")

#list
x = ["apple", "banana", "cheery"]
print(x)
print(type(x))
print("\n")

#tuple
x = ("appple", "banana", "cherry")
print(x)
print(type(x))
print("\n")

#set
x = {"apple", "banana", "cherry"}
print(x)
print(type(x))
print("\n")

#dict
x = {"name" : "John", "age" : 36}
print(x)
print(type(x))
print("\n")

#frozenset
x = frozenset({"apple", "banana", "cherry"})
print(x)
print(type(x))
print("\n")

#range
x = range(6)
print(x)
print(type(x))
print("\n")

#bool
x = True
print(x)
print(type(x))
print("\n")

#bytes
x = b"Hello"
print(x)
print(type(x))
print("\n")

#bytearray
x = bytearray(5)
print(x)
print(type(x))
print("\n")

#memoryview
x = memoryview(bytes(5))
print(x)
print(type(x))
print("\n")

#NoneType
x = None
print(x)
print(type(x))
print("\n")
```

Float can be a scientific number with an "e" to indicate power of 10

```python
x = 35e3
y = 12E4
z = -87.7e100

print(x)
print(type(x))
print(y)
print(type(y))
print(z)
print(type(z))
```
